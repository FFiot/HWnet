```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.layers import Dense
```

```python
import sys
sys.path.append('../HWnet_base/')
from util import plt_scatter

from HWnet_evaluate import HWnet_evaluate
from HWnet_plus_keras_layer import HWnet_plus
```

```python
def test_fucntion(x):
    return np.sin(x**2 * np.pi * 4)/2 - (x)**2 + 0.5

data_train = (np.random.random(4096).astype(np.float32).reshape((-1,1)) - 0.5)*2
# data_train = np.linspace(-1.0, 1.0, 101).astype(np.float32).reshape((-1,1))
target_train = test_fucntion(data_train).reshape((-1,1))
plt_scatter(data_train,y_true=target_train)
```

```python
evaluate = HWnet_evaluate(0.2, vector_dims=1)


class lossStop(keras.callbacks.Callback):
    def __init__(self):
        super(lossStop, self).__init__()
    
    def on_train_begin(self, logs={}):
        self.loss_history = np.array([])
        
    def on_epoch_end(self, epoch, logs={}):
        self.loss_history = np.append(self.loss_history, logs['loss'])
        if len(self.loss_history) >= 20 and self.loss_history[-10:].std() <= 0.001:
            self.model.stop_training = True


for epotch in range(10):
    idx = evaluate.push_array(data_train[:, 0])

    net = HWnet_plus(evaluate.get_parameter())
    
    model = keras.Sequential([net, Dense(1)])
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.1), loss=tf.keras.losses.Huber())
    his = model.fit(data_train, target_train, batch_size=128, epochs=100, verbose=0, callbacks=[lossStop()])

    print('********************','Train_cnt=%d'%len(his.history['loss']) ,'Eoptch=%d'%(epotch+1), 'loss=%0.6f'%(his.history['loss'][-1]), '********************')
    
    predict = model(data_train)
    plt_scatter(data_train, y_predict=predict, y_true=target_train)

    loss = (predict - target_train)
    loss = np.abs(loss)

    evaluate.set_vector(net.get_vector())
    evaluate.feed_loss(idx, loss)
    print(evaluate.dataFrame())

    evaluate.evaluate_split()
    
#     print(evaluate.dataFrame())
```
